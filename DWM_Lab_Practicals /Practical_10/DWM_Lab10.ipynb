```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA

# Load dataset
df = pd.read_csv('IRIS - IRIS.csv')
print(df.head())
# Remove label column for unsupervised learning
X = df.drop('species', axis=1)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

# Map original species to numeric labels (for comparison)
label_encoder = LabelEncoder()
df['species_encoded'] = label_encoder.fit_transform(df['species'])

# Use PCA to reduce features to 2D for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plotting clusters
plt.figure(figsize=(10, 5))

# Clustered data
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['cluster'], cmap='rainbow')
plt.title("K-Means Clustered Data")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")

# Actual species
plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['species_encoded'], cmap='rainbow')
plt.title("Actual Species")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA

# ✅ Create a short custom dataset
data = {
    'feature1': [1.4, 1.3, 1.5, 4.5, 4.7, 4.6, 6.7, 6.5, 6.6],
    'feature2': [0.2, 0.2, 0.3, 1.5, 1.6, 1.4, 2.3, 2.2, 2.4],
    'feature3': [5.1, 4.9, 5.0, 6.2, 6.4, 6.3, 7.6, 7.8, 7.7],
    'feature4': [3.5, 3.0, 3.2, 2.8, 2.9, 2.7, 3.1, 3.2, 3.0],
    'species': ['setosa', 'setosa', 'setosa', 'versicolor', 'versicolor', 'versicolor', 'virginica', 'virginica', 'virginica']
}

df = pd.DataFrame(data)

# Unsupervised Learning: drop target column
X = df.drop('species', axis=1)

# K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

# Encode species to compare with clusters
label_encoder = LabelEncoder()
df['species_encoded'] = label_encoder.fit_transform(df['species'])

# PCA for 2D visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualization
plt.figure(figsize=(10, 5))

# K-Means Clusters
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['cluster'], cmap='rainbow')
plt.title("K-Means Clustered Data")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")

# Actual Species
plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['species_encoded'], cmap='rainbow')
plt.title("Actual Species")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")

plt.tight_layout()
plt.show()
```
